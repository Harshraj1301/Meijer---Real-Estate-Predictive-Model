```python
import pandas as pd

# Read the CSV files into dataframes
final_nearby_places_df = pd.read_csv('final_nearby_places_df.csv')
final_sentiments_df = pd.read_csv('final_sentiments_dataframe.csv')

# Remove the word "County" from the 'county' column in final_nearby_places_df
final_nearby_places_df['County'] = final_nearby_places_df['County'].str.replace(' County', '')

master_df = pd.merge(final_nearby_places_df, final_sentiments_df, on='County', how='left')
master_df.drop('county', axis=1, inplace=True)

# Save the resultant dataframe to a new CSV file
master_df.to_csv('master_df.csv', index=False)
```
